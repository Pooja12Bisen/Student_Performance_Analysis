```python
sns.boxplot(df["Hours_Studied"])
plt.show()
```

```python
# showing boxplot all columns
for col in continuous_columns:
    plt.figure(figsize=(7, 5))
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'Boxplot of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Value', fontsize=12)
    plt.grid(True, axis='x', linestyle='--', alpha=0.7)
    plt.show()
```

```python
#showing histplot for all continuous coloumn
for col in continuous_columns:
    plt.figure(figsize=(7, 5))
    sns.histplot(data=df, x=col, bins=20, kde=True, color='skyblue')
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```

```python
# Example: Barplot showing average Exam_Score by Gender (seaborn)
plt.figure(figsize=(7, 5))
sns.barplot(x='Gender', y='Exam_Score', data=df)
plt.title('Average Exam Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Exam Score')
plt.show()
```

```python
columns = ['Hours_Studied', 'Attendance', 'Previous_Scores', 'Sleep_Hours', 'Motivation_Level']

for col in columns:
    plt.figure()
    sns.scatterplot(x=col, y='Exam_Score', data=df)
    plt.title(f"{col} vs Exam_Score")
    plt.show()
```

```python
correlation = df['Hours_Studied'].corr(df['Exam_Score'])
print(f'Correlation between Hours_Studied and Exam_Score: {correlation}')
```
